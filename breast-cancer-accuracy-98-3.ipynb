# Read Data

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.pandas.set_option('display.max_rows',None) 
pd.pandas.set_option('display.max_columns',None) 
```

```python
#load dataset
df = pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')
df.head()
```

```python
#datafame shape
print(df.shape)
```

```python
#observing the central tendency and other essentials of the dataset
df.describe()
```

Great news! The data has no null values. Also all variables except *diagnosis* are in float form.

# PREPROCESS

We will not be needing these.

```python
#unnamed 32 is not essential so dropping it
df = df.drop(['Unnamed: 32', 'id'], axis=1)
df.head()
```

```python
#counting the number of Malignant and Benign in diagnosis column
df['diagnosis'].value_counts()
```

There are two diagnosis results and they are in object form. I will replace M with 1 and B with 0. So if the *diagnosis* is 1, which means it is cancerous.
* M, malignant: 1, Cancerous
* B, benign: 0, Not Cancerous

```python
#encoding the the target feature
df['diagnosis']= df['diagnosis'].replace('M', 1)
df['diagnosis']= df['diagnosis'].replace('B', 0)
```

```python
df.head()
```

```python
#plotting the corellation matrix
corr = df.corr()
plt.figure(figsize=(18,18))
sns.heatmap(corr, cmap='coolwarm', annot = True)
plt.show()
```

```python
#finding out the positively corelated feature
cc=corr[abs(corr['diagnosis']) > 0.5].index
print('- Number of most correlated features = ', len(cc))
print('--------------------------------------------------')
print('- Most correlated features is: \n ',cc)
```

```python
acc=df[df.columns[:]].corr()['diagnosis']
print('All features  with thier correlations is: \n',acc)
```

```python
#finding out the negatively corelated feature
cc2=corr[abs(corr['diagnosis']) <= 0.5].index
print('- Number of Least correlated features = ', len(cc2))
print('--------------------------------------------------')
print('- Least correlated features is: \n ',cc2)
```

```python
#selecting 20 most essential features
df = df[['diagnosis', 'radius_mean', 'area_mean',
       'compactness_mean', 'concavity_mean', 'concave points_mean',
       'perimeter_worst', 'area_worst', 'compactness_worst',
       'concavity_worst', 'concave points_worst','texture_mean', 
         'smoothness_mean', 'symmetry_mean','area_se','fractal_dimension_se',
         'texture_worst', 'smoothness_worst', 'symmetry_worst',
       'fractal_dimension_worst']]
df.head()
```

```python
print(df.shape)
df.tail(10)
```

```python
# create hist and kde plots to observe the data distribution
fig, ax = plt.subplots(ncols=5, nrows=4, figsize=(30,20))
index = 0
ax = ax.flatten()

for col, value in df.items():
    col_dist = sns.histplot(value, ax=ax[index], color="blue", label="100% Equities", kde=True, stat="density", linewidth=0)
    col_dist.set_xlabel(col,fontsize=18)
    col_dist.set_ylabel('density',fontsize=18)
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
# selecting only the skewed columns to transform it in gaussian distribution
df_new = df[['radius_mean', 'area_mean',
       'compactness_mean', 'concavity_mean', 'concave points_mean',
       'area_worst', 'compactness_worst',
       'concavity_worst', 'area_se','fractal_dimension_se',
         'symmetry_worst', 'fractal_dimension_worst']].copy()
print(df_new.shape)
df_new.head()
```

```python
#using standard deviation method to treat the skewness
for feature in df_new:
    upper_limit = df_new[feature].mean() + 3*df_new[feature].std()
    lower_limit = df_new[feature].mean() - 3*df_new[feature].std()
    df_new[feature] = np.where(df_new[feature]>upper_limit, upper_limit,
                    np.where( df_new[feature]<lower_limit, lower_limit,
                    df_new[feature]))
```

```python
# create hist and kde plots to observe the data distribution
fig, ax = plt.subplots(ncols=4, nrows=3, figsize=(20,10))
index = 0
ax = ax.flatten()

for col, value in df_new.items():
    scol_dist = sns.histplot(value, ax=ax[index], color="blue", label="100% Equities", kde=True, stat="density", linewidth=0)
    col_dist.set_xlabel(col,fontsize=18)
    col_dist.set_ylabel('density',fontsize=18)
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
#merging the normalized and the remaining datframes into one for making the model
df_to_be_added = df[['diagnosis', 'perimeter_worst', 'concave points_worst','texture_mean', 
         'smoothness_mean', 'symmetry_mean', 'texture_worst', 'smoothness_worst']].copy()

frames = [df_to_be_added, df_new]

df_final = pd.concat(frames, axis=1)
df_final.head()
```

```python
df_final.shape
```

```python
# create hist and kde plots to observe the data distribution
fig, ax = plt.subplots(ncols=5, nrows=4, figsize=(30,20))
index = 0
ax = ax.flatten()

for col, value in df_final.items():
    col_dist = sns.histplot(value, ax=ax[index], color="blue", label="100% Equities", kde=True, stat="density", linewidth=0)
    col_dist.set_xlabel(col,fontsize=18)
    col_dist.set_ylabel('density',fontsize=18)
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
#plotting the corellation matrix
corr = df_final.corr()
plt.figure(figsize=(12,6))
sns.heatmap(corr, cmap='coolwarm', annot = True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import metrics
```

```python
#dividing the dataframe into training and target features
x = df_final.drop(['diagnosis'], axis=1)
y = df_final['diagnosis']
```

```python
x.shape, y.shape
```

```python
#splitting the dataframe and keeping 80% of the data for training and rest 20% for testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify = y)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# LOGISTIC REGRESSION

```python
#training the data using Logistic Regression Classifier
from sklearn.linear_model import LogisticRegression
classifier_lr = LogisticRegression()
classifier_lr.fit(x_train, y_train)

prediction_lr = classifier_lr.predict(x_test)
print(classification_report(y_test, prediction_lr, 
                            target_names = ['Malignant(Class 0)','Benign (Class 1)']))
metrics.plot_roc_curve(classifier_lr, x_test, y_test)
```

```python
#plotting the confusion matrix for test set
from sklearn.metrics import plot_confusion_matrix, accuracy_score, classification_report
plot_confusion_matrix(classifier_lr,x_test,y_test)
plt.show()  
```

```python
#printing the accuracy for test set
print('Accuracy of Logistic Regression model is {}'.format(accuracy_score(y_test,prediction_lr)*100))
```

# DECISION TREE

```python
#training the data using Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(x_train, y_train)
prediction_dt = tree.predict(x_test)
print(classification_report(y_test, prediction_dt))
metrics.plot_roc_curve(tree, x_test, y_test)
```

```python
#plotting the confusion matrix for test set
from sklearn.metrics import confusion_matrix
plot_confusion_matrix(tree,x_test,y_test)
plt.show()
```

```python
#printing the accuracy for test set
print('Accuracy of Decision Tree model is {}'.format(accuracy_score(y_test,prediction_dt)*100))
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
forest.fit(x_train, y_train)
prediction_rf = forest.predict(x_test)
print(classification_report(y_test, prediction_rf))
metrics.plot_roc_curve(forest, x_test, y_test)
```

```python
#plotting the confusion matrix for test set
plot_confusion_matrix(forest,x_test,y_test)
plt.show()
```

```python
#printing the accuracy for test set
print('Accuracy of Random Forest model is {}'.format(accuracy_score(y_test,prediction_rf)*100))
```

# XGBOOST

```python
import xgboost
xgb = xgboost.XGBClassifier(earning_rate=0.009, n_estimators=350, subsample=0.8, max_depth=7)
xgb.fit(x_train,y_train)
prediction_xgb = xgb.predict(x_test)
print(classification_report(y_test, prediction_xgb, 
                            target_names = ['Malignant(Class 0)','Benign (Class 1)']))
metrics.plot_roc_curve(xgb, x_test, y_test)
```

```python
#plotting the confusion matrix for test set
plot_confusion_matrix(xgb,x_test,y_test)
plt.show()
```

```python
#printing the accuracy for test set
print('Accuracy of XGB model is {}'.format(accuracy_score(y_test,prediction_xgb)*100))
```
